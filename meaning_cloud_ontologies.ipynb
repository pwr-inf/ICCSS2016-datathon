```python
import requests

url = "http://api.meaningcloud.com/class-1.1"
headers = {'content-type': 'application/x-www-form-urlencoded'}

iptc = []

for row_id, row in speeches.iterrows():
    payload = "key=9b45d20a6a1e99a054fb10c23e0b1695&txt={}&model=IPTC_en".format(row['Texts'])
    response = requests.request("POST", url, data=payload, headers=headers)
    iptc.append(simplejson.loads(response.text))
```

```python
speeches_pkl['ontology'] = iptc
```
