```python
import pickle
import pandas as pd
from  datetime import date
import timestring
```

```python
speeches=pickle.load(open('corelation_speeches.p','rb'))
```

```python
fb_comments=pickle.load(open('corelation_facebook_comments.p','rb'))
```

```python
fb_posts=pickle.load(open('corelation_facebook.p','rb'))
```

```python
twitter=pickle.load(open('corelation_twitter.p','rb'))
```

```python
twitter[2][0]=list(twitter[2][0])
twitter[2][1]=list(twitter[2][1])
twitter[2][2]=list(twitter[2][2])
```

```python
data=[speeches,fb_comments,fb_posts,twitter]
def to_frame(d):
    
    f1=pd.DataFrame(list([d[0][0],d[2][0]]),list([d[1][0],'date']),dtype=float).transpose()
    f2=pd.DataFrame(list([d[0][1],d[2][1]]),list([d[1][1],'date']),dtype=float).transpose()
    f3=pd.DataFrame(list([d[0][2],d[2][2]]),list([d[1][2],'date']),dtype=float).transpose()
    return f1.merge(f2,how='outer').merge(f3,how='outer').fillna(0)
frames=map(to_frame,data)
```

```python
frames
```

```python
merged=reduce(lambda x,y:x.merge(y,how='outer',on='date'),frames).interpolate().fillna(0)
merged['date']=merged['date'].apply(timestring.Date)
```

```python
mergedF=merged[merged['date']>  datetime(2015, 4, 23,0,0,0)]
mergedF
```

```python

mergedF=mergedF[mergedF['date']<  datetime(2016, 5, 23,0,0,0)].drop('date',axis=1)
mergedF
```

```python
corr=mergedF.corr()
corr.insert(0, 'from', corr.index)
```

```python
corr
```

```python
pd.melt(corr, id_vars='from', value_vars=list(corr.columns[1:]),
            var_name='to', value_name='p').to_csv('correlation_tab.csv')
```
